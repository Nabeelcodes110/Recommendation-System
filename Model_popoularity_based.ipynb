```python
import pandas as pd
import csv
```

```python
dataset = pd.read_csv('kurtiData.csv')
dataset.head()
```

```python
num_rating_df = dataset.groupby('product_id').count()['rating'].reset_index()
num_rating_df
```

```python
avg_rating_df = dataset.groupby('product_id').mean()['rating'].reset_index()
avg_rating_df
```

```python
popular_df = avg_rating_df[num_rating_df['rating']>=250].sort_values('rating',ascending=False).head(50)
# popular_df = popular_df.merge(dataset,on='product_id')[['user_id', 'product_id', 'product_url', 'image_url', 'price', 'rating']]
popular_df = popular_df.merge(dataset,on='product_id').drop_duplicates('product_id')[['product_id','rating_y', 'image_url', 'price' ]]
popular_df
```

```python
import pickle
pickle.dump(popular_df,open('popular.pkl','wb'))
```
